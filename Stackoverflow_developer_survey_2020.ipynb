```python
#importing libraries and reading csv data set
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
filename = r"/home/iddie/Desktop/python/header_footer/survey_results_public.csv"
pd.set_option("display.max_columns",61)
df = pd.read_csv(filename)
df
```

```python
#grouping datasets into countries
grouped = df.groupby('Country')
grouped.get_group('Kenya')
```

```python
#number of respondents of each country
no_respondents = df['Country'].value_counts()
no_respondents
```

```python
#number of python programmers in each country
pythonistas=grouped['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
pythonistas
```

```python
#Concatenating two colns to create a new dataframe
py_df = pd.concat([no_respondents,pythonistas],axis='columns',sort=False)
```

```python
py_df
```

```python
#Renaming colums of the new dataframe
py_df.rename(columns={'Country':'Num of respondents','LanguageWorkedWith':'Pythonistas'},inplace=True)
```

```python
py_df.sort_values(by=['Num of respondents'], ascending=False)
```

```python
#Top 10 countries with most developers
sns.countplot(y=df['Country'], order=df['Country'].value_counts().index[:10])
```

```python
#percentage of python coders using simple created function and use of apply()
def perc(x,y):
    return x/y*100
py_df['Percentage'] = py_df.apply(lambda x: perc(x['Pythonistas'],x['Num of respondents']), axis=1)
```

```python
py_df
```

```python
py_df['percentage2'] = (py_df['Pythonistas']/py_df['Num of respondents'])*100
py_df
```
